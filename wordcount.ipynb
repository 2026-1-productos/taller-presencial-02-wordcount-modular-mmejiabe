```python

import os  
```

```python
def read_files_from_folder(input_folder):
    return os.listdir(input_folder)
```

```python
def process_input_files(input_folder, files):
    counter = {}
    for file in files:
        path = os.path.join(input_folder, file)
        counter = process_file(path, counter)
    return counter
```

```python
def process_file(path, counter):
    with open(path, "r", encoding="utf-8") as file:
        for line in file:
            counter = process_line(line, counter)
    return counter
```

```python
def process_line(line, counter):
    for word in line.split():
        word = word.lower().strip(",.!?")
        counter[word] = counter.get(word, 0) + 1
    return counter
```

```python
def create_output_folder(output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
```

```python
def save_results(counter, filename):
    with open(filename, "w", encoding="utf-8") as f:
        for key, value in counter.items():
            f.write(f"{key}\t{value}\n")
```

```python
def main(input_folder, output_folder):
    files = read_files_from_folder(input_folder)
    counter = process_input_files(input_folder, files)
    create_output_folder(output_folder)
    save_results(counter, filename="data/output/results.tsv")
```

```python
main(
    input_folder="data/input/",
    output_folder="data/output",
)
```
